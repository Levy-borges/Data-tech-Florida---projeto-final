```python
# Importando as bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
# Carregar e analisar os dados inicialmente

df = pd.read_csv('data-join.csv')
df.head(1)
```

```python
df.dtypes
```

```python
# Supondo que seu dataframe esteja na variável df

# 1. Converter 'date' para datetime
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 2. Converter colunas numéricas (int)
int_cols = ['items_delivered', 'items_missing', 'age', 'Trips', 'customer_age']
for col in int_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

# 3. Converter 'order_amount' para float (você já tem float64, mas manteremos exemplo)
if 'order_amount' in df.columns:
    df['order_amount'] = (
        df['order_amount']
        .astype(str)
        .replace('[\$,]', '', regex=True)
    )
    df['order_amount'] = pd.to_numeric(df['order_amount'], errors='coerce')

# 4. Converter 'price' para float (caso haja símbolos de dólar, vírgulas etc.)
if 'price' in df.columns:
    df['price'] = (
        df['price']
        .astype(str)
        .replace('[\$,]', '', regex=True)
    )
    df['price'] = pd.to_numeric(df['price'], errors='coerce')

# 5. Converter 'delivery_hour' (se quiser extrair apenas a hora de "HH:MM:SS")
#    do contrário, se for um número inteiro (0-23), converta direto:
if 'delivery_hour' in df.columns:
    # Se estiver no formato HH:MM:SS, converta e extraia a hora
    df['delivery_hour'] = pd.to_datetime(df['delivery_hour'], format="%H:%M:%S", errors='coerce').dt.hour
    # Se quiser manter como int
    df['delivery_hour'] = df['delivery_hour'].fillna(0).astype(int)

# 6. Converter colunas textuais (object) que não requerem cálculo
#    Ajuste a lista conforme necessário
str_cols = [
    'order_id','region','driver_id','customer_id','driver_name',
    'customer_name','missing_product_type','product_id','product_name','category'
]
for col in str_cols:
    if col in df.columns:
        df[col] = df[col].astype(str)

# Exibir resultado
print("\nTipos de dados após conversão:\n", df.dtypes)
print("\nPreview dos dados:\n", df.head())
```

```python
# Se a coluna existir, removemos
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Verificar estatísticas
print(df['items_delivered'].describe())

# Encontrar registros suspeitos: <= 0 (negativos ou zero)
suspeitos = df[df['items_delivered'] <= 0]
if not suspeitos.empty:
    print("\nRegistros com 'items_delivered' <= 0:")
    print(suspeitos)

# Visualizar distribuição com boxplot

sns.boxplot(x=df['items_delivered'])
plt.title("Boxplot - Items Delivered")
plt.show()
```

```python
# Estatísticas descritivas
print(df['items_missing'].describe())

# Identificar registros incoerentes:
# 1) items_missing < 0 (impossível ter itens faltantes negativos)
# 2) items_missing > items_delivered (não deveria faltar mais itens do que o total entregue)
incoerentes = df[
    (df['items_missing'] < 0) |
    (df['items_missing'] > df['items_delivered'])
]

if not incoerentes.empty:
    print("\nRegistros incoerentes em 'items_missing':")
    print(incoerentes)

# Visualizar distribuição com boxplot

sns.boxplot(x=df['items_missing'])
plt.title("Boxplot - Items Missing")
plt.show()
```

```python
# Ver estatísticas
print(df['age'].describe())

# Definir intervalo plausível (ex.: entre 18 e 80, conforme política de motoristas)
age_min, age_max = 18, 70

# Localizar valores fora do intervalo
incoerentes_age = df[
    (df['age'] < age_min) | 
    (df['age'] > age_max)
]

if not incoerentes_age.empty:
    print("\nRegistros incoerentes em 'age':")
    print(incoerentes_age)

# Boxplot para ver distribuição de 'driver_age'

sns.boxplot(x=df['age'])
plt.title("Boxplot - Driver Age")
plt.show()
```

```python
# Defina as colunas numéricas que deseja inspecionar
num_cols = ['items_delivered', 'items_missing', 'Trips', 'age', 'customer_age']

# Plotar um boxplot para cada coluna
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot - {col}")
    plt.show()
```

```python
# 1. Agrupar por driver_id somando itens_missing e items_delivered
driver_stats = df.groupby('driver_id').agg({
    'items_missing': 'sum',
    'items_delivered': 'sum'
}).reset_index()

# 2. Calcular a taxa de reclamação (ex.: itens_missing / itens_delivered)
# Evitar divisão por zero usando np.where ou fillna
driver_stats['driver_complaint_rate'] = np.where(
    driver_stats['items_delivered'] == 0,
    0,  # ou algum valor para sinalizar problema?
    driver_stats['items_missing'] / driver_stats['items_delivered']
)

# 3. Ordenar e visualizar os top 10 motoristas com maior taxa
driver_stats_top10 = driver_stats.sort_values(
    by='driver_complaint_rate',
    ascending=False
).head(10)

print("Top 10 Motoristas com maior taxa de reclamação (items_missing/items_delivered):")
print(driver_stats_top10)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(
    x='driver_id', 
    y='driver_complaint_rate', 
    data=driver_stats_top10, 
    palette='Reds'
)
plt.title("Top 10 Motoristas - Taxa de Reclamação")
plt.xlabel("Driver ID")
plt.ylabel("Taxa de Reclamação (items_missing / items_delivered)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Exemplo: consideramos estas features (ajuste conforme o dataset)
features = ['driver_complaint_rate', 'Trips', 'age']

# Criar um subset do dataframe
X = driver_stats.merge(df[['driver_id', 'age', 'Trips']].drop_duplicates(), 
                       on='driver_id', how='left')[features]

# Opcional: normalizar (para que todas as colunas fiquem na mesma escala)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Instanciar o modelo (defina 'contamination' como fração de outliers esperada)
iso_forest = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
iso_forest.fit(X_scaled)

# Prever anomalias (-1 = anomalia, 1 = normal)
driver_stats['anomaly_label'] = iso_forest.predict(X_scaled)
driver_stats['anomaly_score'] = iso_forest.decision_function(X_scaled)

# Selecionar motoristas anômalos
anomalias = driver_stats[driver_stats['anomaly_label'] == -1]

print("Motoristas marcados como anômalos:")
print(anomalias)
```

```python
# 1. Crie um subset apenas com motoristas suspeitos (outliers)
suspect_drivers = driver_stats[driver_stats['anomaly_label'] == -1]['driver_id'].unique()

# 2. Filtre o dataframe original (df) para pegar apenas pedidos feitos por esses motoristas
suspect_orders = df[df['driver_id'].isin(suspect_drivers)].copy()

# 3. Verifique algumas estatísticas relevantes
print("Quantidade de pedidos feitos pelos motoristas suspeitos:", suspect_orders.shape[0])

# 3.1. Análise por região
region_analysis = suspect_orders.groupby('region').agg({
    'items_missing': 'sum',          # total de itens faltantes
    'items_delivered': 'sum',        # total de itens entregues
    'order_amount': 'mean'           # valor médio do pedido
}).reset_index()

region_analysis['missing_rate'] = region_analysis['items_missing'] / region_analysis['items_delivered']
region_analysis.sort_values(by='missing_rate', ascending=False, inplace=True)

print("\nAnálise por Região (motoristas suspeitos):")
print(region_analysis)

# 3.2. Análise por categoria de produto
# Se sua base estiver em formato "largo" (product_id_1, category_1 etc.), 
# você pode precisar “despivotar” ou tratar cada product_id_x. 
# Aqui assumimos que cada linha tem colunas product_name e category:
category_analysis = suspect_orders.groupby('category').agg({
    'items_missing': 'sum',
    'items_delivered': 'sum',
    'order_amount': 'mean'
}).reset_index()

category_analysis['missing_rate'] = category_analysis['items_missing'] / category_analysis['items_delivered']
category_analysis.sort_values(by='missing_rate', ascending=False, inplace=True)

print("\nAnálise por Categoria de Produto (motoristas suspeitos):")
print(category_analysis)

# 3.3. Análise de order_amount em geral
print("\nEstatísticas de order_amount para pedidos de motoristas suspeitos:")
print(suspect_orders['order_amount'].describe())

# 4. Se quiser, ver detalhes de produtos mais afetados (exemplo):
# agrupar por product_name
if 'product_name' in suspect_orders.columns:
    product_analysis = suspect_orders.groupby('product_name').agg({
        'items_missing': 'sum',
        'items_delivered': 'sum',
        'order_amount': 'mean'
    }).reset_index()
    
    product_analysis['missing_rate'] = product_analysis['items_missing'] / product_analysis['items_delivered']
    product_analysis.sort_values(by='missing_rate', ascending=False, inplace=True)

    print("\nProdutos mais afetados (entre motoristas suspeitos):")
    print(product_analysis.head(10))
```
